About Dataset:
The Sustainable Development Report 2023 dataset includes thorough information about sustainability and advancements
made in relation to the Sustainable Development Goals (SDGs) for a number of nations.
The dataset's entries each include statistics on a country's sustainability rankings,
regional placement, and SDG performance. This dataset offers insightful information about
international sustainability initiatives and enables a nuanced evaluation of nations' development
towards sustainability.


Dataset:

```python
import pandas as pd

df = pd.read_csv('sdg_index.csv',index_col="year")

# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Find the mean value for each goal score
mean_values = df[goal_columns].mean()

# Create a list to store dictionaries with goal, mean value, and countries below the mean
below_mean_data = []

# Iterate through each goal score column
for goal in goal_columns:
    # Mean value for the current goal
    mean_value = df[goal].mean()  
    # Concatenate unique country names into a string
    countries_below_mean = ', '.join(df.loc[df[goal] < mean_value, 'country'].unique())  
    below_mean_data.append({'Goal': goal, 'Mean_Value': mean_value, 'Countries_Below_Mean': countries_below_mean})

#DataFrame from the list of dictionaries
below_mean_df = pd.DataFrame(below_mean_data)

# Print the results row by row
for index, row in below_mean_df.iterrows():
    print(f"Goal: {row['Goal']}\nMean Value: {row['Mean_Value']}\nCountries Below Mean: {row['Countries_Below_Mean']}\n{'='*50}\n")

# Save the DataFrame to a CSV file
below_mean_df.to_csv('below_mean_values_countries.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sdg_index.csv',index_col="year")

# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Find the mean value for each goal score
mean_values = df[goal_columns].mean()

# List to store the number of countries below the mean for each goal
countries_below_mean_count = []

# Iterate through each goal score column
for goal in goal_columns:
    mean_value = df[goal].mean()  # Mean value for the current goal
    num_countries_below_mean = df[df[goal] < mean_value]['country'].nunique()  # Number of countries below the mean
    countries_below_mean_count.append({'Goal': goal, 'Num_Countries_Below_Mean': num_countries_below_mean})

# DataFrame from the list of dictionaries
below_mean_count_df = pd.DataFrame(countries_below_mean_count)

# Print the results row by row
for index, row in below_mean_count_df.iterrows():
    print(f"Goal: {row['Goal']}\nNumber of Countries Below Mean: {row['Num_Countries_Below_Mean']}\n{'='*50}\n")

# Save the DataFrame to a CSV file
below_mean_count_df.to_csv('below_mean_count.csv', index=False)

# Bar chart for the number of countries below the mean for each goal
plt.figure(figsize=(12, 6))
plt.bar(below_mean_count_df['Goal'], below_mean_count_df['Num_Countries_Below_Mean'], color='lightcoral')
plt.xlabel('SDG Goals')
plt.ylabel('Number of Countries Below Mean')
plt.title('Number of Countries Below Mean for Each SDG Goal')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 141, 5)) 
plt.savefig('NumberofCountriesBelowMeanforEachSDGGoal.pdf', bbox_inches='tight')
plt.show()
```
