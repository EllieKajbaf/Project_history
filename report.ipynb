```python
"""
This code calculates mean and standard deviation scores for specified SDGs,
and then prints the summary statistics.

"""
import pandas as pd

df = pd.read_csv('sdg_index.csv',index_col="year")

# Increase the maximum number of rows displayed
pd.set_option('display.max_rows', None)

# Select the columns to calculate statistics for
selected_columns = ['goal_1_score', 'goal_2_score',
                    'goal_3_score', 'goal_4_score',
                    'goal_5_score',
                    'goal_6_score', 'goal_7_score', 
                    'goal_8_score', 'goal_9_score', 
                    'goal_10_score',
                    'goal_11_score', 'goal_12_score', 
                    'goal_13_score', 'goal_14_score',
                    'goal_15_score',
                    'goal_16_score', 'goal_17_score']

# Group the data by the 'country' column
grouped_data = df.groupby('country')

# Initialize an empty DataFrame to store the summary statistics
summary_stats = pd.DataFrame()

# Iterate through the selected columns and calculate the mean and standard deviation for each region
for column in selected_columns:
    mean_col_name = f'{column}_mean'
    std_col_name = f'{column}_std'
    summary_stats[mean_col_name] = grouped_data[column].mean()
    summary_stats[std_col_name] = grouped_data[column].std()

# Display the summary statistics for the selected columns for each region
print(summary_stats)
```

```python
"""
This code calculates the median and maximum values for
specified goal scores for each country,
and then prints the summary statistics.

"""
import pandas as pd

# Increase the maximum number of rows displayed
pd.set_option('display.max_rows', None)

df = pd.read_csv('sdg_index.csv',index_col="year")

# Select the columns you want to calculate statistics for
selected_columns = ['goal_1_score', 'goal_2_score', 
                    'goal_3_score', 'goal_4_score', 
                    'goal_5_score','goal_6_score', 
                    'goal_7_score', 'goal_8_score', 
                    'goal_9_score', 'goal_10_score',
                    'goal_11_score', 'goal_12_score', 
                    'goal_13_score', 'goal_14_score', 
                    'goal_15_score','goal_16_score',
                    'goal_17_score']

# Initialize an empty DataFrame to store the median and max values
summary_stats = pd.DataFrame()

# Group the data by 'country' and calculate the median and max for each selected column
for column in selected_columns:
    grouped_data = df.groupby('country')[column].agg(['median', 'max'])
    summary_stats[column + '_median'] = grouped_data['median']
    summary_stats[column + '_max'] = grouped_data['max']

# Display the summary statistics for each country and goal
print(summary_stats)
```

```python
"""
This code categorizes countries into three groups ('Wealthy,' 'Poor,' and 'Middle') based on their 'sdg_index_score'
for each year in the range 2000 to 2023.
The categorization is done using quantiles (specifically, the 10th and 90th percentiles). 
The mean 'sdg_index_score' and the set of countries falling into each category are calculated and 
stored in separate DataFrames (wealthy_df, poor_df, and middle_df). 
The results are then displayed.

"""
import pandas as pd

file_path = 'sdg_index.csv'
data = pd.read_csv('sdg_index.csv', index_col="country_code")
# Initialize sets to store unique country names 
#for each category for each year
wealthy_countries_by_year = []
poor_countries_by_year = []
middle_countries_by_year = []
wealthy_means = []
poor_means = []
middle_means = []

# Iterate through years and calculate means 
# and get country names for each category
years = list(range(2000, 2023))

for year in years:
    column_name = 'sdg_index_score'
    column_data = data[data['year'] == year][column_name]
    column_quantile_10 = column_data.quantile(0.10)
    column_quantile_90 = column_data.quantile(0.90)
   
    wealthy_data = data[(data['year'] == year) &
                        (data[column_name] > column_quantile_90)]
    wealthy_data = data[(data['year'] == year) &
                        (data[column_name] > column_quantile_90)]
    poor_data = data[(data['year'] == year) & 
                     (data[column_name] < column_quantile_10)]
    middle_data = data[(data['year'] == year) &
                       (data[column_name] >= column_quantile_10) &
                       (data[column_name] <= column_quantile_90)]
    
    wealthy_mean = wealthy_data[column_name].mean()
    poor_mean = poor_data[column_name].mean()
    middle_mean = middle_data[column_name].mean()
    
    wealthy_means.append(wealthy_mean)
    poor_means.append(poor_mean)
    middle_means.append(middle_mean)


    wealthy_countries_by_year.append(set(wealthy_data['country']))
    poor_countries_by_year.append(set(poor_data['country']))
    middle_countries_by_year.append(set(middle_data['country']))
    
# Create DataFrames to display the results
wealthy_df = pd.DataFrame({'Year': years,
                           'Wealthy_Mean': wealthy_means,
                           'Wealthy_Countries': wealthy_countries_by_year})
poor_df = pd.DataFrame({'Year': years,
                        'Poor_Mean': poor_means,
                        'Poor_Countries':poor_countries_by_year})
middle_df = pd.DataFrame({'Year': years, 
                          'Middle_Mean': middle_means, 
                          'Middle_Countries': middle_countries_by_year})

# Set 'Year' as the index for each DataFrame
wealthy_df.set_index('Year', inplace=True)
poor_df.set_index('Year', inplace=True)
middle_df.set_index('Year', inplace=True)

    # Display the results
with pd.option_context('display.max_colwidth', None):
    print(f"Results for {file_path}:")
    print("Wealthy Countries:")
    print(wealthy_df)

    print("\nPoor Countries:")
    print(poor_df)

    print("\nMiddle Countries:")
    print(middle_df)
```

```python
"""
This code calculates the mean value of the 'sdg_index_score' column for each year and
categorizes countries as either above or below the mean for each year.
The results are displayed in two separate DataFrames, 
one for above mean countries (Above_mean_countries_df) and 
one for below mean countries (Below_mean_countries_df).

"""

import pandas as pd

data = pd.read_csv('sdg_index.csv', index_col="country_code")

file_path = 'sdg_index.csv'

# Initialize sets to store unique country names for each category for each year
above_mean_countries_by_year = []
below_mean_countries_by_year = []
means_value_by_year = []


# Iterate through years and calculate means and get country names for each category
years = list(range(2000, 2023))
#sdg_index_score : The overall sustainability (SDG index) score of the country.
for year in years:
    column_name = 'sdg_index_score'
    column_data = data[data['year'] == year][column_name]
    mean_value_by_year = column_data.mean()   
   
    above_mean_countries_data = data[(data['year'] == year) & 
                                     (data[column_name] >  mean_value_by_year)]
    below_mean_countries_data = data[(data['year'] == year) & 
                                     (data[column_name] <  mean_value_by_year)]    
    
    means_value_by_year.append(mean_value_by_year)

    above_mean_countries_by_year.append(set(above_mean_countries_data['country']))
    below_mean_countries_by_year.append(set(below_mean_countries_data['country']))
    
    
# Create DataFrames to display the results
Above_mean_countries_df = pd.DataFrame({'Year': years, 
                                        'Mean_values_by_year': means_value_by_year,
                                        'Above_mean_Countries': above_mean_countries_by_year})
Below_mean_countries_df = pd.DataFrame({'Year': years, 
                                        'Mean_values_by_year': means_value_by_year,
                                        'below_mean_Countries': below_mean_countries_by_year})


# Set 'Year' as the index for each DataFrame
Above_mean_countries_df.set_index('Year', inplace=True)
Below_mean_countries_df.set_index('Year', inplace=True)

# Increase the maximum number of items displayed in a sequence
pd.set_option('display.max_seq_items', None)

# Display the results
with pd.option_context('display.max_colwidth', None):
    print(f"Results for {file_path}:")
    print("Above mean countries:")
    print(Above_mean_countries_df)

    print("\n Below mean countries_df:")
    print(Bellow_mean_countries_df)
```
