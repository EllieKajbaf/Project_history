```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
This code finds, prints, and visualizes the maximum values for each SDG goal, 
including associated countries, and saves the results to a CSV file.

"""

import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate

df = pd.read_csv('sdg_index.csv',index_col="year")
# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Find the maximum value for each goal score
max_values = df[goal_columns].max()

# List to store dictionaries with goal, max value, and countries
max_data = []

# Iterate through each goal score column
for goal in goal_columns:
    # Maximum value for the current goal
    max_value = df[goal].max()  
     # Concatenate unique country names
    countries_with_max_value = ', '.join(df.loc[df[goal] == max_value, 'country'].unique()) 
    max_data.append({'Goal': goal, 'Max_Value': max_value, 'Countries_with_Max_Value': countries_with_max_value})

# Create a DataFrame from the list of dictionaries
max_df = pd.DataFrame(max_data)

# Print the results row by row
for index, row in max_df.iterrows():
    print(f"Goal: {row['Goal']}\nMax Value: {row['Max_Value']}\nCountries with Max Value:
          {row['Countries_with_Max_Value']}\n{'='*50}\n")

# Save the DataFrame to a CSV file
max_df.to_csv('max_values_countries.csv', index=False)

# Visualization 1: Bar chart for maximum values
plt.figure(figsize=(12, 6))
plt.bar(max_df['Goal'], max_df['Max_Value'])
plt.xlabel('SDG Goals')
plt.ylabel('Maximum Value')
plt.title('Maximum Value for Each SDG Goal')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 101, 5)) 
plt.savefig('MaximumValueforEachSDGGoal.pdf', bbox_inches='tight')
plt.show()

# Visualization 2: Table for associated countries
for index, row in max_df.iterrows():
    table_data = {'Goal': [row['Goal']],
                  'Max Value': [row['Max_Value']],
                  'Countries with Max Value': [row['Countries_with_Max_Value']]}
    table = tabulate(table_data, headers='keys', tablefmt='pretty')
    print(table)
    print('='*50)
```
