**Generating descriptive statistics of a DataFrame:
df.describe() provides a summary of the central tendency, dispersion, and shape of the distribution of a dataset.
The statistics include:**

Count: Number of non-null values.
Mean: Mean (average) value.
std: Standard deviation, which measures the amount of variation or dispersion of a set of values.
min: Minimum value.
25%: First quartile (25th percentile).
50%: Median (50th percentile).
75%: Third quartile (75th percentile).
max: Maximum value.
