```python
"""
This code is for identifying the minimum values for each goal score in the DataFrame, 
along with the countries that achieved these minimum values.
The results are printed and saved to a CSV file for further analysis.

"""
import pandas as pd
df = pd.read_csv('sdg_index.csv',index_col="year")
# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Find the minimum value for each goal score
min_values = df[goal_columns].min()

# List to store dictionaries with goal, min value, and countries
min_data = []

# Iterate through each goal score column
for goal in goal_columns:
    min_value = df[goal].min()  # Minimum value for the current goal
    countries_with_min_value = ', '.join(df.loc[df[goal] == min_value, 'country'].unique())  # Concatenate unique country names into a string
    min_data.append({'Goal': goal, 'Min_Value': min_value, 'Countries_with_Min_Value': countries_with_min_value})

#DataFrame from the list of dictionaries
min_df = pd.DataFrame(min_data)

# Print the results row by row
for index, row in min_df.iterrows():
    print(f"Goal: {row['Goal']}\nMin Value: {row['Min_Value']}\nCountries with Min Value: {row['Countries_with_Min_Value']}\n{'='*50}\n")

# Save the DataFrame to a CSV file
min_df.to_csv('min_values_countries.csv', index=False)
```

```python
"""
These visualizations and tables provide a overview of the minimum values 
for each SDG goal and the associated countries with those minimum values.
The bar chart visually represents the minimum values, while the tables provide detailed information in a tabular format.
"""

import matplotlib.pyplot as plt
from tabulate import tabulate

# Assuming min_df contains the results

# Visualization 1: Bar chart for minimum values
plt.figure(figsize=(12, 6))
plt.bar(min_df['Goal'], min_df['Min_Value'])
plt.xlabel('SDG Goals')
plt.ylabel('Minimum Value')
plt.title('Minimum Value for Each SDG Goal')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 101, 5)) 
plt.savefig('MinimumValueforEachSDGGoal.pdf', bbox_inches='tight')
plt.show()

# Visualization 2: Table for associated countries
for index, row in min_df.iterrows():
    table_data = {'Goal': [row['Goal']],
                  'Min Value': [row['Min_Value']],
                  'Countries with Min Value': [row['Countries_with_Min_Value']]}
    table = tabulate(table_data, headers='keys', tablefmt='pretty')
    print(table)
    print('='*50)
```
