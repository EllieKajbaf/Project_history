```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sdg_index.csv',index_col="year")
```

```python
df.info()
```

```python
'''
This code computes and organizes statistical summary information (mean and standard deviation) 
for each goal score in the original DataFrame and presents it in a new DataFrame.
'''

import pandas as pd

# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Calculate the mean and standard deviation for each goal score
mean_values = df[goal_columns].mean()
std_values = df[goal_columns].std()

# Create a new DataFrame to store the results
result_df = pd.DataFrame({'Goal': mean_values.index, 'Mean': mean_values.values, 'Std': std_values.values})

# Print the result_df 
print(result_df)
```

```python
"""
This code visualizes mean values for different SDG Goals using blue bars and includes error bars (in red) 
to represent the standard deviation.
The resulting plot provides insights into the distribution of scores for each SDG Goal.
"""
import matplotlib.pyplot as plt

# the result_df contains the mean and std values as calculated in the previous code

# Set the size of the figure
plt.figure(figsize=(12, 12))

# Create a bar plot for mean values
plt.bar(result_df['Goal'], result_df['Mean'], color='blue', alpha=0.7, label='Mean')

# Add error bars for standard deviation
plt.errorbar(result_df['Goal'], result_df['Mean'], yerr=result_df['Std'], fmt='none', ecolor='red', capsize=5, label='Std Dev')

# Set plot labels and title
plt.xlabel('SDG Goals')
plt.ylabel('Score')
plt.title('Mean and Standard Deviation for Each SDG Goal')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.savefig('MeanStandardDeviationforEachSDGGoal.pdf', bbox_inches='tight')
# Show the plot
plt.show()
```
