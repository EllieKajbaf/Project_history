```python
"""
This code calculates the mean value for each goal score column in the DataFrame and then identifies countries
that have scores above the mean for each goal. The results are printed and saved to a CSV file.
"""

import pandas as pd

df = pd.read_csv('sdg_index.csv',index_col="year")
# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Find the mean value for each goal score
mean_values = df[goal_columns].mean()

# List to store dictionaries with goal, mean value, and countries above mean
above_mean_data = []

# Iterate through each goal score column
for goal in goal_columns:
    mean_value = df[goal].mean()  # Mean value for the current goal
    countries_above_mean = ', '.join(df.loc[df[goal] > mean_value, 'country'].unique())  # Concatenate unique country names into a string
    above_mean_data.append({'Goal': goal, 'Mean_Value': mean_value, 'Countries_Above_Mean': countries_above_mean})

# Create a DataFrame from the list of dictionaries
above_mean_df = pd.DataFrame(above_mean_data)

# Print the results row by row
for index, row in above_mean_df.iterrows():
    print(f"Goal: {row['Goal']}\nMean Value: {row['Mean_Value']}\nCountries Above Mean: {row['Countries_Above_Mean']}\n{'='*50}\n")

# Save the DataFrame to a CSV file
above_mean_df.to_csv('above_mean_countries.csv', index=False)
```

```python
"""
This code calculates the mean value for each goal score column in the 'df' and 
then determines the number of countries with scores above the mean for each goal. 

"""

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sdg_index.csv',index_col="year")
# Select only the columns containing goal scores
goal_columns = df.columns[3:]

# Find the mean value for each goal score
mean_values = df[goal_columns].mean()

# List to store the number of countries above the mean for each goal
countries_above_mean_count = []

# Iterate through each goal score column
for goal in goal_columns:
    mean_value = df[goal].mean()  # Mean value for the current goal
    num_countries_above_mean = df[df[goal] > mean_value]['country'].nunique()  # Number of countries above the mean
    countries_above_mean_count.append({'Goal': goal, 'Num_Countries_Above_Mean': num_countries_above_mean})

# DataFrame from the list of dictionaries
above_mean_count_df = pd.DataFrame(countries_above_mean_count)

# Print the results row by row
for index, row in above_mean_count_df.iterrows():
    print(f"Goal: {row['Goal']}\nNumber of Countries Above Mean: {row['Num_Countries_Above_Mean']}\n{'='*50}\n")

# Save the DataFrame to a CSV file
above_mean_count_df.to_csv('above_mean_count.csv', index=False)

# Bar chart for the number of countries above the mean for each goal
plt.figure(figsize=(12, 6))
plt.bar(above_mean_count_df['Goal'], above_mean_count_df['Num_Countries_Above_Mean'], color='skyblue')
plt.xlabel('SDG Goals')
plt.ylabel('Number of Countries Above Mean')
plt.title('Number of Countries Above Mean for Each SDG Goal')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 141, 5)) 
plt.savefig('NumberofCountriesAboveMeanforEachSDGGoal.pdf', bbox_inches='tight')
plt.show()
```
