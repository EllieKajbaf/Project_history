```python
"""
This code calculates mean and standard deviation scores for specified SDGs, and then prints the summary statistics.

"""
import pandas as pd

df = pd.read_csv('sdg_index.csv',index_col="year")

# Increase the maximum number of rows displayed
pd.set_option('display.max_rows', None)

# Select the columns to calculate statistics for
selected_columns = ['goal_1_score', 'goal_2_score', 'goal_3_score', 'goal_4_score', 'goal_5_score',
                    'goal_6_score', 'goal_7_score', 'goal_8_score', 'goal_9_score', 'goal_10_score',
                    'goal_11_score', 'goal_12_score', 'goal_13_score', 'goal_14_score', 'goal_15_score',
                    'goal_16_score', 'goal_17_score']

# Group the data by the 'country' column
grouped_data = df.groupby('country')

# Initialize an empty DataFrame to store the summary statistics
summary_stats = pd.DataFrame()

# Iterate through the selected columns and calculate the mean and standard deviation for each region
for column in selected_columns:
    mean_col_name = f'{column}_mean'
    std_col_name = f'{column}_std'
    summary_stats[mean_col_name] = grouped_data[column].mean()
    summary_stats[std_col_name] = grouped_data[column].std()

# Display the summary statistics for the selected columns for each region
print(summary_stats)
```

```python
"""
This code calculates the median and maximum values for specified goal scores for each country,
and then prints the summary statistics.

"""
import pandas as pd

# Increase the maximum number of rows displayed
pd.set_option('display.max_rows', None)

df = pd.read_csv('sdg_index.csv',index_col="year")

# Select the columns you want to calculate statistics for
selected_columns = ['goal_1_score', 'goal_2_score', 'goal_3_score', 'goal_4_score', 'goal_5_score',
                    'goal_6_score', 'goal_7_score', 'goal_8_score', 'goal_9_score', 'goal_10_score',
                    'goal_11_score', 'goal_12_score', 'goal_13_score', 'goal_14_score', 'goal_15_score',
                    'goal_16_score', 'goal_17_score']

# Initialize an empty DataFrame to store the median and max values
summary_stats = pd.DataFrame()

# Group the data by 'country' and calculate the median and max for each selected column
for column in selected_columns:
    grouped_data = df.groupby('country')[column].agg(['median', 'max'])
    summary_stats[column + '_median'] = grouped_data['median']
    summary_stats[column + '_max'] = grouped_data['max']

# Display the summary statistics for each country and goal
print(summary_stats)
```

```python
"""
This code calculates the minimum and maximum values for specified goal scores for each country,
and then prints the summary statistics.

"""
import pandas as pd

# Increase the maximum number of rows displayed
pd.set_option('display.max_rows', None)

df = pd.read_csv('sdg_index.csv',index_col="year")

# Select the columns you want to calculate statistics for
selected_columns = ['goal_1_score', 'goal_2_score', 'goal_3_score', 'goal_4_score', 'goal_5_score',
                    'goal_6_score', 'goal_7_score', 'goal_8_score', 'goal_9_score', 'goal_10_score',
                    'goal_11_score', 'goal_12_score', 'goal_13_score', 'goal_14_score', 'goal_15_score',
                    'goal_16_score', 'goal_17_score']

# Initialize an empty DataFrame to store the median and max values
summary_stats = pd.DataFrame()

# Group the data by 'country' and calculate the min and max for each selected column
for column in selected_columns:
    grouped_data = df.groupby('country')[column].agg(['min', 'max'])
    summary_stats[column + '_min'] = grouped_data['min']
    summary_stats[column + '_max'] = grouped_data['max']

# Display the summary statistics for each country and goal
print(summary_stats)
```

```python
"""
 This code reads data from a CSV file, filters it to select data for a specific country ('Canada'), 
 and then creates and displays a bar plot using seaborn to visualize the 'No Poverty' index for each year.

"""
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('sdg_index.csv',index_col="year")
# Filter the DataFrame to select a specific country ('Canada')
country_name = 'Canada'
Canada_data = df[df['country'] == country_name]
df_Canada = pd.DataFrame(Canada_data)
plt.figure(figsize=(20,10))
sns.barplot(df_Canada,x='year', y='goal_1_score')
plt.xticks(rotation=90)
plt.ylabel('No Poverty index')
plt.show()
```
